# PCA 

    (Principal Component Analysis) is an unsupervised learning technique and a commonly used dimensionality reduction method. It involves mapping the features of data from an original dimensionality of n to a lower dimensionality of k (where k < n) in a way that minimizes information loss.


## Create data 
Create a set of two-dimensional data

```python
import numpy as np
import matplotlib.pyplot as plt

# create data, n points
n = 20
x = np.linspace(0, n-1, n)
y = 3*x+2
#x=np.array([2.5,0.5,2.2,1.9,3.1,2.3,2,1,1.5,1.1])
#y=np.array([2.4,0.7,2.9,2.2,3,2.7,1.6,1.1,1.6,0.9])
# Introduce random disturbance.
x += np.random.rand(x.shape[0]) *n/3
y += np.random.rand(y.shape[0]) *n/3

# Normalization
x /= n
y /= n

# Data Creation done，find mean and normalize


xmean0 = x - x.mean()
ymean0 = y - y.mean()

# Process matrix.
data=np.matrix([[xmean0[i],ymean0[i]] for i in range(n)])

# Compute the covariance matrix
## solution 1
cov1 = data.T*data/(n-1)
## solution 2
cov2 = np.cov(xmean0,ymean0)

#cov1-cov2

# find eigenvalue and eigenvector

eig_val, eig_vec = np.linalg.eig(cov1)

# rank the eigenvalue
order = np.argsort(eig_val)

# Select the top K eigenvectors corresponding to the modulus of  eigenvalues.
K = 0
eig_vec_K = eig_vec[order==K]
reducedata = eig_vec[order==K]*data.T
reducedata = np.array(reducedata)


# Recover the data using all components
fcdata = eig_vec * data.T
fcdata = np.array(fcdata)

plt.plot(np.zeros(n),reducedata[0,:],'+',label="1st Component")
plt.plot(fcdata[0,:], fcdata[1,:],'ro',label="PCA data")
plt.plot(xmean0,ymean0,'b.',label="Origine data(Mean0)")
plt.title("Principal Components Analysis")
plt.ylim()
plt.legend()
plt.grid()
```

```python
K = 0
eig_vec[order>K] = 0.
reducedata = eig_vec*data.T
reducedata = np.array(reducedata)
plt.plot(reducedata[0,:],np.zeros(n),'+',label="1st Component")
plt.plot(fcdata[0,:], fcdata[1,:],'ro',label="PCA data")
plt.plot(xmean0,ymean0,'b.',label="Origine data(Mean0)")
plt.title("Principal Components Analysis")
plt.ylim()
plt.legend()
plt.grid()
```
